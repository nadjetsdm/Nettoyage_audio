```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Charger le fichier audio
audio_path = "C:/Users/nsidimou/Downloads/mon_projet_perso/audio.mp3"
signal, sr = librosa.load(audio_path, sr=None)

# 1. Afficher la forme d'onde
time = np.linspace(0, len(signal) / sr, num=len(signal))

plt.figure(figsize=(12, 4))
plt.plot(time, signal, color="blue")
plt.title("Forme d'onde")
plt.xlabel("Temps (secondes)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Détection des outliers
threshold = 3 * np.std(signal)  # Définir un seuil basé sur l'écart-type
outliers = np.where(np.abs(signal) > threshold)[0]  # Indices des outliers

plt.figure(figsize=(12, 4))
plt.plot(time, signal, label="Signal", color="blue")
plt.scatter(
    time[outliers],
    signal[outliers],
    color="red",
    label="Outliers",
    zorder=5,
)
plt.title("Détection des Outliers")
plt.xlabel("Temps (secondes)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
